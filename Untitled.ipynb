```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(1,20,100)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('Time')
plt.show()
```

```python
x=np.linspace(1,20,100)
y=np.sin(x)
z=np.fft.fft(y)
plt.subplot(1,2,1)
plt.plot(x,y,'r',label='EEG')
plt.xlabel('Time')
plt.legend()
plt.show()

plt.subplot(1,2,2)
plt.plot(y,abs(z))
plt.xlabel('Time')
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,3))
ax[0].plot(x,y)
ax[0].set_title('EEg')

ax[1].plot(x,z)
ax[1].set_title('FFT')
```
